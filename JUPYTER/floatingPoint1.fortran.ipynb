```fortran
write(6,*) "Check cancellation: give three number "
a =  7.0
b = -a
c =  0.333333

aa = 7.0
bb = -aa 
cc =  0.333333333333333333
```

```fortran
write(6,*) "Check a,b,c    =", a,b,c
write(6,*) "Check aa,bb,cc =", aa,bb,cc
```

```fortran
write(6,*) "Check (a+b)+c = ", ((a+b)+c) 
write(6,*) "Check a+(b+c) = ", (a+(b+c))
write(6,*) "Check (a+c)+b = ", ((a+c)+b)
!
write(6,*) "Check (aa+bb)+cc = ", ((aa+bb)+cc) 
write(6,*) "Check aa+(bb+cc) = ", (aa+(bb+cc))
write(6,*) "Check (aa+cc)+bb = ", ((aa+cc)+bb)
```

```fortran
write(6,*) "Error % (a+b)+c = ", 100*(((a+b)+c)-c)/c, "%"
write(6,*) "Error % a+(b+c) = ", 100*((a+(b+c))-c)/c, "%"
write(6,*) "Error % (a+c)+b = ", 100*(((a+c)+b)-c)/c, "%"
!
write(6,*) "Error % (aa+bb)+cc = ", 100*(((aa+bb)+cc)-cc)/cc, "%" 
write(6,*) "Error % aa+(bb+cc) = ", 100*((aa+(bb+cc))-cc)/cc, "%"
write(6,*) "Error % (aa+cc)+bb = ", 100*(((aa+cc)+bb)-cc)/cc, "%"
```
