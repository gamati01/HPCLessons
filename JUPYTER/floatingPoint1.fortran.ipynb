```fortran
write(6,*) "Check a =", a
write(6,*) "Check b =", b
write(6,*) "Check c =", c
```

```fortran
write(6,*) "Check (a+b)+c = ", ((a+b)+c) 
write(6,*) "Check a+(b+c) = ", (a+(b+c))
write(6,*) "Check (a+c)+b = ", ((a+c)+b)
```

```fortran
write(6,*) "Error (a+b)+c = ", (((a+b)+c)-c)/c 
write(6,*) "Error a+(b+c) = ", ((a+(b+c))-c)/c
write(6,*) "Error (a+c)+b = ", (((a+c)+b)-c)/c
```
