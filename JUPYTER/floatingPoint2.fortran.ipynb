```fortran
real    :: sum1, sum2 
integer :: i, n, iter
real    :: pi=3.14159265359
!
```

```fortran
! forward sum..
    n = 500
    write(6,*) "Forward sum "
    do iter =1, 11
       sum1 = 0
       do i=1, n*iter, 1. 
    	   sum1=sum1+(1.0/(float(i)*float(i)))
       end do
       write(6,*) 'Elements =', n*iter, "PI=", sqrt(6.0*sum1), "Error =", sqrt(6.0*sum1)-pi
	end do
!
```

```fortran
! backward sum
    write(6,*) "Backward sum "
    n = 50000
    do iter =1, 10
       sum2 = 0
       do i=n*iter, 1, -1
    	   sum2=sum2+(1.0/(float(i)*float(i)))
	   end do
       write(6,*) 'Elements =', n*iter, "PI=", sqrt(6.0*sum2), "Error =", sqrt(6.0*sum2)-pi
    enddo
```
