```fortran
integer, parameter:: nx=4096
integer, parameter:: ny=4096
integer :: i,j,k
!
real, dimension(nx,ny) :: a ! matrix (origin)
real, dimension(nx,ny) :: b ! matrix (origin)
real, dimension(nx,ny) :: c ! matrix (destination)
real :: time1, time2 ! timing
real :: time3, time4 ! timing
!
```

```fortran
write(6,*) "Matrix size", nx, ny
```

```fortran
write(6,*) "Matrix initialization"
do i = 1, nx
do j = 1, ny
   a(i,j) = float(i)/float(j)
   b(i,j) = float(j)/float(i)
   c(i,j) = 0.0
enddo
enddo
```

```fortran
write(6,*) "Check", a(nx,ny), b(nx,ny), c(nx,ny)
```

```fortran
! product #1
call system("date")
do i = 1, nx
do j = 1, ny
do k = 1, nx
   c(i,j) = c(i,j) + a(i,k)*b(k,j)
enddo 
enddo
enddo
call system("date")

```

```fortran
write(6,*) "Check", c(2,7)
```

```fortran
! product #2
call system("date")
do j = 1, ny
do k = 1, nx
do i = 1, nx
   c(i,j) = c(i,j) + a(i,k)*b(k,j)
enddo 
enddo
enddo
call system("date")
```
